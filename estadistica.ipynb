```python
import numpy as np # vectores y matrices
import pandas as pd # tablas
import matplotlib.pyplot as plt #graficar
import statistics #operaciones basica de estadistica
from scipy.stats import norm #estadistica avanzada y matematica
```

```python
# en este caso estoy utilizando numpy
notasAlumnos = [1,2,3,4,5,6,7,9,10,10,10,10,10,10,10,10,10,10,10,10,10,10,10,10]
promedio = np.mean(notasAlumnos)
varianza = np.var(notasAlumnos)
desviacionEstandar = np.std(notasAlumnos)
mensaje = "media es: {}, la variaza es: {} y desviacion estandar es: {}"
print(mensaje.format(promedio,varianza,desviacionEstandar))
```

```python
intervalos = range(min(notasAlumnos), max(notasAlumnos)+1)
print(intervalos)
plt.hist(notasAlumnos,bins=intervalos, color='purple', rwidth=0.98)
plt.scatter(promedio,0, color="red", linewidths=5)
plt.title("histograma")
plt.xlabel("Notas")
plt.ylabel("frecuencia")
plt.xticks(intervalos)
```

```python
serie = [10,25,32,48,49,45,22,46,29,49,43,48,15,45,62,38,43,69,38,46,34,54,45,46,18,45,32,34,62,44,42,44,62,42,35,66,12,27,49,35,55,37,31,64,17,33,80,68,42,43,61,74,14,34,76,28,26,34,38,26]
promedio1 = np.mean(serie)
varianzaEstandar1 = np.std(serie)

plt.axvline(x=promedio1+varianzaEstandar1, color='blue', linestyle='-', label='Promedio')
plt.axvline(x=promedio1-varianzaEstandar1, color='blue', linestyle='-', label='Promedio')
plt.axvline(x=promedio1, color='blue', linestyle='-', label='Promedio')

plt.hist(serie,bins=6, color='purple')
plt.title("histograma")
plt.xlabel("cantidad")
plt.ylabel("frecuencia")

print(promedio1)
print(varianzaEstandar1)
```
